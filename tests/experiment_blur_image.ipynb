```python
import cv2
import numpy as np
from PIL import Image
```

```python
def blur_image(image):
    w, h, _ = image.shape
    center = (int(w/2), int(h/2))    
    blurred_img = cv2.GaussianBlur(image, (21, 21), 0, cv2.BORDER_REFLECT)
    mask = np.zeros((w, h, 3), dtype=np.uint8)
    mask = cv2.circle(mask, center, 100, (255, 255, 255), -1)
    out = np.where(mask==(255, 255, 255), image, blurred_img)
    return out

image_list = [
    './test_images_processed/fortune-mation.portrait.jpg',
    './test_images_processed/head-of-an-old-man.portrait.jpg',
    './test_images_processed/madame-cezanne-in-blue-1890.portrait.jpg'
]
for i in image_list:
    img = cv2.cvtColor(cv2.imread(i), cv2.COLOR_BGR2RGB)
    print('original')
    display(Image.fromarray(img))
    img = blur_image(img)
    print('blur')
    display(Image.fromarray(img))
```
